# floating point number:
$mantissa * base^{exponent}$
- normalise: $mantissa\in[1,base)$
- expressible numbers: [overflow, expressable, underflow(positive negative), expressable, verflow ]
- binary float: $0.01101 = 2^{-2} + 2^{-3} + 2^{-5}$
- floating point has inexact result due to rounding and truncating
- when comparing, we chose a epsilon to check whether a is in the neighbourhood of b
- IEEE (1bit sign + 8bit exponent + 23bit significand)
- denormalised: when the exponent is all zero, all one exponent for inf and nan

# gates:

# time dependent circuit:
at first, the nodes are unknown(if not given), then get the next timestep
# flipflop:
P and Q are unknown if both input are 1, otherwist P+S = 1,Q+R=1

if R=S=1,the circuit is stable when P$\neq$Q,unstable otherwise with P=Q always

```python
import schemdraw
from schemdraw import logic
with schemdraw.Drawing() as d:
    gS = d.add(logic.Nand())
    d.move_from(gS.in1,dy = -3)
    gR = d.add(logic.Nand())
    d.move_from(gS.in1)
    d.add(logic.Line().left(1).label('S'))
    d.move_from(gR.in2)
    d.add(logic.Line().left(1).label('R'))
    d.move_from(gS.out)
    d.add(logic.Line().right(1).label('P'))
    d.move_from(gR.out)
    d.add(logic.Line().right(1).label('Q'))
    d.move_from(gS.in2)
    d.add(logic.Wire('N').to(gR.out))
    d.move_from(gR.in1)
    d.add(logic.Wire('N').to(gS.out))
```
